## Day 4-5：Pandas 练习

```python
import pandas as pd
import seaborn as sns

# 📖 加载 Titanic 数据集
df = sns.load_dataset("titanic")
print(f"数据大小: {df.shape}")
print(f"列名: {df.columns.tolist()}")
```

```python
# 📖 数据概览
print("前5行:")
print(df.head())
print("\n数据类型:")
print(df.info())
print("\n统计摘要:")
print(df.describe())
```

```python
# ⚡ 练习1：女性平均年龄是多少？男性呢？
print(f"女性平均年龄: {df[df['sex']=='female']['age'].mean():.1f}")
print(f"男性平均年龄: {df[df['sex']=='male']['age'].mean():.1f}")
```

```python
# ⚡ 练习2：按性别分组，计算存活率和平均票价
result = df.groupby("sex").agg({
    "survived": "mean",
    "fare": "mean",
    "age": "mean"
})
print(result)
```

```python
# ⚡ 练习3：年龄大于30的乘客，存活率是多少？
print(f"30岁以上存活率: {df[df['age']>30]['survived'].mean():.2%}")
print(f"30岁以下存活率: {df[df['age']<=30]['survived'].mean():.2%}")
```

```python
# ⚡ 练习4：头等舱中，男性和女性的存活率分别是多少？
first_class = df[df["pclass"] == 1]
print(first_class.groupby("sex")["survived"].mean())
```

```python
# ⚡ 练习5：缺失值检查
print("缺失值数量:")
print(df.isna().sum())
print("\n缺失比例:")
print(df.isna().mean() * 100)
```

```python
# ⚡ 练习6：用中位数填充年龄的缺失值，再验证是否还有缺失
df["age"].fillna(df["age"].median(), inplace=True)
print(f"填充后年龄缺失值: {df['age'].isna().sum()}")
```

## Day 6：数据可视化

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")

df = sns.load_dataset("titanic")
df["age"].fillna(df["age"].median(), inplace=True)
```

```python
# 📖 看：直方图 — 年龄分布
plt.figure(figsize=(10, 5))
sns.histplot(df["age"], bins=30, kde=True)
plt.title("年龄分布")
plt.xlabel("年龄")
plt.ylabel("人数")
plt.show()
```

```python
# ⚡ 画：柱状图 — 各舱位存活率
survival_by_class = df.groupby("pclass")["survived"].mean()
plt.figure(figsize=(8, 5))
plt.bar(survival_by_class.index, survival_by_class.values, color=["red", "blue", "green"])
plt.title("各舱位存活率")
plt.xlabel("舱位等级")
plt.ylabel("存活率")
plt.xticks([1, 2, 3])
plt.ylim(0, 1)
plt.show()
```

```python
# ⚡ 画：箱线图 — 幸存者和遇难者的年龄分布
plt.figure(figsize=(8, 5))
sns.boxplot(x="survived", y="age", data=df)
plt.title("幸存 vs 遇难 年龄分布")
plt.show()
```

```python
# ⚡ 画：热力图 — 各因素相关性
corr_cols = ["age", "fare", "pclass", "survived", "sibsp", "parch"]
corr = df[corr_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("相关性热力图")
plt.show()

# 思考：哪些因素和存活率相关性最强？
```

## Day 7：完整 EDA

```python
# 📖 完整 EDA 模板
# 把之前学到的所有技巧整合到一起

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")

# 加载数据
df = sns.load_dataset("titanic")

print("="*50)
print("1. 数据概览")
print("="*50)
print(f"行数: {df.shape[0]}, 列数: {df.shape[1]}")
print(f"\n前5行:\n{df.head()}")
print(f"\n数据类型:\n{df.dtypes}")

print("\n" + "="*50)
print("2. 缺失值处理")
print("="*50)
print(f"缺失比例:\n{df.isna().mean() * 100}")
df["age"].fillna(df["age"].median(), inplace=True)
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)
if "deck" in df.columns:
    df.drop("deck", axis=1, inplace=True)

print("\n" + "="*50)
print("3. 关键统计")
print("="*50)
print(f"总体存活率: {df['survived'].mean():.2%}")
print(f"女性存活率: {df[df['sex']=='female']['survived'].mean():.2%}")
print(f"男性存活率: {df[df['sex']=='male']['survived'].mean():.2%}")
print(f"头等舱存活率: {df[df['pclass']==1]['survived'].mean():.2%}")
print(f"三等舱存活率: {df[df['pclass']==3]['survived'].mean():.2%}")
```

```python
# 可视化部分
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

sns.countplot(x="survived", data=df, ax=axes[0, 0])
axes[0, 0].set_title("存活情况")

sns.countplot(x="sex", hue="survived", data=df, ax=axes[0, 1])
axes[0, 1].set_title("性别 × 存活")

sns.countplot(x="pclass", hue="survived", data=df, ax=axes[0, 2])
axes[0, 2].set_title("舱位 × 存活")

sns.histplot(df["age"], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("年龄分布")

sns.boxplot(x="survived", y="fare", data=df[df["fare"] < 200], ax=axes[1, 1])
axes[1, 1].set_title("票价 × 存活")

sns.heatmap(df[["age", "fare", "pclass", "survived"]].corr(),
            annot=True, cmap="coolwarm", center=0, ax=axes[1, 2])
axes[1, 2].set_title("相关系数热力图")

plt.tight_layout()
plt.show()
```
