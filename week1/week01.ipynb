```python
#  2.3 函数
# 定义函数
def greet(name):
    """给某人打招呼"""  # 这是文档字符串，解释函数用途
    return f"你好,{name}!"

# 调用函数
msg = greet("小明")
print(msg)

# 多个参数
def add(a,b):
    return a + b

result = add(3,5)  #result = 8

#默认参数
def say_hello(name,title="同学"):
    print(f"{name}{title},你好！")

say_hello('小明')
say_hello('小明','老师')
```

```python
### 2.4 列表推导式
# 传统写法
squares = []
for i in range(10):
    squares.append(i*i)
print()

# 列表推导式 —— 一行搞定
squares = [i*i for i in range(10)]
print(squares)
print()

# 加条件 
even_squares = [i * i for i in range(10) if i % 2 == 0]
print(even_squares)
```

```python
# 2.5异常处理
# 防止程序崩溃
try:
    num = int(input("请输入一个数字："))
    result = 10 / num
    print(f"10 / {num} = {result}")
except ValueError:
    print("你输入的不是数字！")
except ZeroDivisionError:
    print("不能除以0！")
finally:
    print("不管怎么样都会执行这一段")
```

```python
# 练习1：写一个函数 is_even(n)，判断n是否为偶数

def is_even(n):
    if n % 2 == 0:
        print(f"你输入的数字 {n}是偶数 ！")
    else:
        print(f"你输入的数字 {n} 不是偶数")

is_even(9)
is_even(12)
```

```python
# ⚡ 练习2：写一个函数，输入一个数字列表，返回所有偶数的平方

def return_suqaresofD(num):
    nums = []
    for i in num:
        if i % 2 == 0:
            nums.append(i*i)
    print(nums)
return_suqaresofD([1,2,3,4,5,6,7,8,9])
```

```python
# ⚡ 练习3：用列表推导式写上面的函数（一行搞定）
def test(nums):
    return [n*n for n in nums if n % 2 ==0 ]

print(test([1,2,3,4,5,6,7,8,9]))
```

```python
# ⚡ 练习4：写一个函数，接收一个字符串列表，返回长度大于3的字符串

def receive_String(strings):
    return [s for s in strings if len(s) > 3]

test = ["1dadac","张行","涨工资","我是糕手"]
print(receive_String(test))
```

```python
import numpy as np  #惯例 导入为np

#创建数组
arr1 = np.array([1,2,3,4,5])  # 一维数组
arr2 = np.array([[1,2],[3,4]]) #二维数组

print(arr1)
print(arr2)

print("~~~~~~~~~~~~~~~~~~~~~~~")

# 常见创建方式
zeros = np.zeros((3,4))  #三行四列的全0矩阵
print(zeros)
print()

ones = np.ones((2,3))   #两行三列的全1矩阵
print(ones)
print()

range_arr = np.arange(10)  #[0,1,2,...,9] 类似range()
print(range_arr)
print()

lin_arr = np.linspace(0,1,5)  #等间隔，五份
print(lin_arr)
```

```python
# 查看数组属性
arr = np.array([[1,2,3],[4,5,6]])

print(arr.shape)   # 两行三列

print(arr.ndim)   # 维度数

print(arr.size)  #元素数，总共多少个元素

print(arr.dtype)  #数据类型





```
